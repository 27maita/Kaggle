```python
train = dset.iloc[:1400]
test = dset.iloc[1400:]
```

```python
categorical = ['smokes', 'gender']
numerical = ['age', 'height_cm', 'weight_kg', 'heart_rate', 'blood_pressure', 'sleep_hours', 'nutrition_quality', 'activity_index']
```

```python
from sklearn.preprocessing import OneHotEncoder

def preprocess(df, categorical, numerical):
    df = df.copy()

    mask = False
    for col in df.columns:
        mask = mask | df[col].isnull()
        dfnulls = df[mask]
    df.drop(colmns = df)
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler

def preprocess(df, categorical, numerical):
    df = df.copy()
    df = df.dropna()

    df['smokes'] = df['smokes'].replace({'no': 0, 'yes': 1, '0': 0, '1': 1})
    df['gender'] = df['gender'].replace({'M': 0, 'F': 1})
    return df
```

```python
train = preprocess(train, categorical, numerical)
test = preprocess(test, categorical, numerical)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
import xgboost as xgb
from sklearn.svm import SVC

knn = KNeighborsClassifier()
lr = LogisticRegression()
rf = RandomForestClassifier()
xgb = xgb.XGBClassifier()
svc = SVC()
```

```python
from sklearn.model_selection import train_test_split

X_train = train.drop(columns = ['is_fit'])
X_test = test.drop(columns = ['is_fit'])

y_train = train['is_fit']
y_test = test['is_fit']
```

```python
knn.fit(X_train, y_train)
lr.fit(X_train, y_train)
rf.fit(X_train, y_train)
xgb.fit(X_train, y_train)
svc.fit(X_train, y_train)
```

```python
knn_preds = knn.predict(X_test)
lr_preds = lr.predict(X_test)
rf_preds = rf.predict(X_test)
xgb_preds = xgb.predict(X_test)
svc_preds = svc.predict(X_test)
```

```python
from sklearn.metrics import f1_score

knn_f1 = f1_score(y_test, knn_preds)
lr_f1 = f1_score(y_test, lr_preds)
rf_f1 = f1_score(y_test, rf_preds)
xgb_f1 = f1_score(y_test, xgb_preds)
svc_f1 = f1_score(y_test, svc_preds)
print(knn_f1, lr_f1, rf_f1, xgb_f1, svc_f1)
```

```python
from sklearn.metrics import accuracy_score

knn_acc = accuracy_score(y_test, knn_preds)
lr_acc = accuracy_score(y_test, lr_preds)
rf_acc = accuracy_score(y_test, rf_preds)
xgb_acc = accuracy_score(y_test, xgb_preds)
svc_acc = accuracy_score(y_test, svc_preds)

print(knn_acc, lr_acc, rf_acc, xgb_acc, svc_acc)
```
