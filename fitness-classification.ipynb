```python
train = dset.iloc[:1400]
test = dset.iloc[1400:]
```

```python
categorical = ['smokes', 'gender']
numerical = ['age', 'height_cm', 'weight_kg', 'heart_rate', 'blood_pressure', 'sleep_hours', 'nutrition_quality', 'activity_index']
```

```python
from sklearn.preprocessing import OneHotEncoder

def preprocess(df, categorical, numerical):
    df = df.copy()

    mask = False
    for col in df.columns:
        mask = mask | df[col].isnull()
        dfnulls = df[mask]
    df.drop(colmns = df)
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler

def preprocess(df, categorical, numerical):
    df = df.copy()
    df = df.dropna()

    df['smokes'] = df['smokes'].replace({'no': 0, 'yes': 1, '0': 0, '1': 1})
        
    for column in categorical:

        encoder = OneHotEncoder(sparse=False, handle_unknown='ignore')
        transformed = encoder.fit_transform(df[[column]])

        col_names = [f"{column}_{cat}" for cat in encoder.categories_[0]]
        one_hot_df = pd.DataFrame(transformed, columns=col_names, index=df.index)

        df = pd.concat([df.drop(columns=[column]), one_hot_df], axis=1)

    scaler = MinMaxScaler()
    df[numerical] = scaler.fit_transform(df[numerical])
        
    return df
```

```python
train = preprocess(train, categorical, numerical)
test = preprocess(test, categorical, numerical)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
import xgboost as xgb
from sklearn.svm import SVC

knn = KNeighborsClassifier(n_neighbors=7, weights='distance', metric='minkowski', p=2)
lr = LogisticRegression(solver='liblinear', penalty='l2', C=1.0, max_iter=1000)
rf = RandomForestClassifier(n_estimators=250, max_depth=10, min_samples_split=5, min_samples_leaf=2, bootstrap=True, random_state=42)
xgb = xgb.XGBClassifier(n_estimators=200, learning_rate=0.05, max_depth=5, subsample=0.9, colsample_bytree=0.8, objective='binary:logistic', random_state=42)
svc = SVC(C=1.0, kernel='rbf', gamma='scale', probability=True, random_state=42)
```

```python
from sklearn.model_selection import train_test_split

X_train = train.drop(columns = ['is_fit'])
X_test = test.drop(columns = ['is_fit'])

y_train = train['is_fit']
y_test = test['is_fit']
```

```python
knn.fit(X_train, y_train)
lr.fit(X_train, y_train)
rf.fit(X_train, y_train)
xgb.fit(X_train, y_train)
svc.fit(X_train, y_train)
```

```python
knn_preds = knn.predict(X_test)
lr_preds = lr.predict(X_test)
rf_preds = rf.predict(X_test)
xgb_preds = xgb.predict(X_test)
svc_preds = svc.predict(X_test)
```

```python
from sklearn.metrics import f1_score

knn_f1 = f1_score(y_test, knn_preds)
lr_f1 = f1_score(y_test, lr_preds)
rf_f1 = f1_score(y_test, rf_preds)
xgb_f1 = f1_score(y_test, xgb_preds)
svc_f1 = f1_score(y_test, svc_preds)
print(knn_f1, lr_f1, rf_f1, xgb_f1, svc_f1)
```

```python
from sklearn.metrics import accuracy_score

knn_acc = accuracy_score(y_test, knn_preds)
lr_acc = accuracy_score(y_test, lr_preds)
rf_acc = accuracy_score(y_test, rf_preds)
xgb_acc = accuracy_score(y_test, xgb_preds)
svc_acc = accuracy_score(y_test, svc_preds)

print(knn_acc, lr_acc, rf_acc, xgb_acc, svc_acc)
```
