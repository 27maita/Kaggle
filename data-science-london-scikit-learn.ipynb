```python
train = pd.read_csv('/kaggle/input/data-science-london-scikit-learn/train.csv')
trainLabels = pd.read_csv('/kaggle/input/data-science-london-scikit-learn/trainLabels.csv')
test = pd.read_csv('/kaggle/input/data-science-london-scikit-learn/test.csv')
```

```python
def name_columns(df):
    k = 1
    cols = list(df.columns)
    for i in cols:
        df.rename(columns={i : f'column_{k}'}, inplace=True)
        k+=1
    return df

train = name_columns(train)
train
```

```python
test = name_columns(test)
test
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(random_state = 28, max_iter = 300)

lr.fit(train, trainLabels)
```

```python
predictions = list(lr.predict(test))
predictions.append(1)
```

```python

submissions = pd.DataFrame({
    'ID' : [i for i in range(1, len(predictions)+1)] ,
    'Solution' : predictions
})
```

```python
submissions.to_csv('submission.csv', index = False)
```
